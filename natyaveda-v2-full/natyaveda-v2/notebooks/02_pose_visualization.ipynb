# NatyaVeda — 02 Pose Visualization
Visualize skeleton overlays on dance video frames.

```python
import sys
sys.path.insert(0, '..')
import cv2
import numpy as np
import matplotlib.pyplot as plt
from src.inference.predictor import SkeletonVisualizer

viz = SkeletonVisualizer(show_hands=True, show_label=True)

# Load a video frame and pose data
video_path = '../data/refined/bharatanatyam/sample_seg000.mp4'
npz_path   = '../data/processed/bharatanatyam/sample_seg000.npz'

cap  = cv2.VideoCapture(video_path)
data = np.load(npz_path)
kpts = data['keypoints']   # [T, 133, 3]

# Show frame 50
cap.set(cv2.CAP_PROP_POS_FRAMES, 50)
ret, frame = cap.read()
cap.release()

if ret:
    annotated = viz.draw(frame, kpts[min(50, len(kpts)-1)],
                         dance_class='bharatanatyam', dance_conf=0.95)
    rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 6))
    plt.imshow(rgb); plt.axis('off')
    plt.title('Skeleton Overlay — Bharatanatyam')
    plt.tight_layout(); plt.show()
```
