```python
import sys
sys.path.insert(0, '..')
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from src.data_collection.dataset_builder import print_dataset_report

# Show dataset stats
print_dataset_report(Path('../data/processed'))
```

```python
# Load a sample and inspect keypoint structure
sample = np.load('../data/processed/bharatanatyam/sample.npz')
print('Keys:', list(sample.keys()))
print('Keypoints shape:', sample['keypoints'].shape)   # [T, 133, 3]
print('Dance form:', sample['dance_form'])
print('Label:', sample['label'])
```

```python
# Plot body keypoint confidence over time
kpts = sample['keypoints']  # [T, 133, 3]
body_conf = kpts[:, :17, 2].mean(axis=1)  # [T] avg body confidence

plt.figure(figsize=(12, 3))
plt.plot(body_conf, label='Body confidence')
plt.axhline(0.4, color='r', linestyle='--', label='Min threshold')
plt.xlabel('Frame'); plt.ylabel('Confidence'); plt.legend()
plt.title('Body Keypoint Confidence Over Time')
plt.tight_layout(); plt.show()
```
